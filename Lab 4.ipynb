```python
import numpy as np
import matplotlib.pyplot as plt

L = 5
Fc = 100
m = 1
a = np.random.uniform(0, 1, L)
tau = np.random.uniform(0, 1, L)
print(a)
print(tau)
plt.stem(tau, a)  # Plot with markers to clearly see the points
plt.axis([0, 1, 0, 1])
plt.xlabel('tau')
plt.ylabel('a')
plt.title('Random values of a against tau')
plt.show()
```

```python
t=np.linspace(0,1,1001)
tstamp=t[1]
fm=2
x=np.sin(2*np.pi*fm*t)
znum = int(np.floor(tau[0]/tstamp))
y=0
y=y+np.insert(a[0]*x[:-znum],0,np.zeros(znum))
plt.plot(t,x,t,y)
```
