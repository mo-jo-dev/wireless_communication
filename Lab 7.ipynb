```python
def pn_sequence_gen(n):
    sequence = [1]
    n_temp = n
    init = []
    i = 0
    
    while i != n_temp-1:
        init.append(0)
        i = i + 1
    init.append(1)
    i = n-1
    
    for j in range((2**n)-2):
        d = init[n-1] ^ init[n-4]
        while i >= 0:
            init[i] = init[i-1]
            i = i - 1
        init[0] = d
        i = n-1
        print("THe value of d is: ",d)
        print("The list after operation is: ",init)
        sequence.append(init[n-1])
    return sequence


pn_sequence_gen(5)
```
