```python
# gama parellel and gama perpendiculat plot
import matplotlib.pyplot as pyt
import sympy as syms
import numpy as np
import math

Er = 2.2
th = np.linspace(0, 90, 90)
GII = (-Er*np.cos(th)+np.sqrt(Er-np.square(np.sin(th))))/(Er*np.cos(th)+np.sqrt(Er-np.square(np.sin(th))))

pyt.plot(th,abs(GII))
pyt.show()
```
