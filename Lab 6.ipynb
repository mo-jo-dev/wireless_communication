```python
import numpy as np
```

```python
def N(i=1, j=2):
    return i**2 + j**2 + i*j

def ChannelCapacity(radius : float):
    area = 7096e6
    k = 1001
    
    capacity = k*(area)/(N() * (6/4) * np.sqrt(3) * (radius**2))
    return capacity
```

```python
R = np.linspace(100,1000,100)
capacity = [ChannelCapacity(r) for r in R]
```

```python
import matplotlib.pyplot as plt
plt.plot(R, capacity)
plt.xlabel('Radius of cell')
plt.ylabel('Channel Capacity')
plt.show()
```

```python
# i = 1, j = 2
n = 4
d = R*np.sqrt(3) * np.sqrt(N())
Pt = (2e-3)*4*np.pi*(d/0.3)**n
```

```python
plt.plot(10*np.log10(Pt), (capacity))
plt.xlabel('Pt')
plt.ylabel('Cap')
plt.show()
```
